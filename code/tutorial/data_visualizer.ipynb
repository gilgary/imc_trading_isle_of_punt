```python
# Create a new dataframe with only the necessary columns
spread_df = prices[['timestamp', 'product','source_file'] + bid_cols + ask_cols].copy()

# Calculate the individual spreads
spread_df['spread_1'] = spread_df['bid_price_1'] - spread_df['ask_price_1']
spread_df['spread_2'] = spread_df['bid_price_2'] - spread_df['ask_price_2']
spread_df['spread_3'] = spread_df['bid_price_3'] - spread_df['ask_price_3']

spread_df
for group in spread_df['product'].unique():
    group_df = spread_df[spread_df['product'] == group]
    print(f"Group: {group}")
    # Aggregate the total amount of spread_1 in each time period
    for src_file, src_df in group_df.groupby('source_file'):

        fig = go.Figure()
        fig.add_trace(go.Scatter(
            x=src_df['timestamp'],
            y=src_df['spread_1'],
            mode='lines+markers',
            name=src_file
        ))
        fig.update_layout(
            title=f"{group} Spread 1 over Time",
            xaxis_title="Timestamp",
            yaxis_title="Spread 1"
        )
        fig.write_html(f"{group}_spread_1_{src_file}.html")
   
```

```python
def run_cross_product_regression(target, day):
    # Filter by day if needed.
    df = spread_df.copy()
    if day != "all":
        df = df[df['source_file'].str.contains(f"day_{day}.csv")]

    # Pivot the dataframe: index is timestamp, columns are products, and values are spread_1.
    pivot_df = df.pivot_table(index='timestamp', columns='product', values='spread_1')

    # Drop rows with missing values.
    pivot_df = pivot_df.dropna()

    if target not in pivot_df.columns:
        print(f"Target product {target} not found in data")
        return

    # Make the target spread the dependent variable; regress on the spreads of the other products.
    y = pivot_df[target]
    X = pivot_df.drop(columns=target)
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    
    print(f"Regression of {target} spread on spreads of other products:")
    print(model.summary())
```

```python
def run_regressions(group, day):
    require 'daru'
    require 'statsample'

    # Assumes that "spread_df" is a Daru::DataFrame with at least the columns:
    # 'timestamp', 'product', 'source_file', 'spread_1', 'spread_2', 'spread_3'
    def run_regressions(group, day)
        # filter by product
        filtered = spread_df.filter_rows { |row| row['product'] == group }
        # filter by day if day is not "all"
        if day != "all"
            filtered = filtered.filter_rows { |row| row['source_file'].include?("day_#{day}.csv") }
        end

        # sort by timestamp to prepare lagged variables
        sorted = filtered.sort_by { |row| row['timestamp'] }

        # extract spread series as arrays
        spread1 = sorted['spread_1'].to_a
        spread2 = sorted['spread_2'].to_a
        spread3 = sorted['spread_3'].to_a

        # --- Contemporaneous Regression ---
        # Regress spread1 on spread2 and spread3 (using all rows)
        contemp_df = Daru::DataFrame.new({
            'spread1' => Daru::Vector.new(spread1),
            'spread2' => Daru::Vector.new(spread2),
            'spread3' => Daru::Vector.new(spread3)
        })
        reg_contemp = Statsample::Regression::Multiple::Normal.new_from_dataset(contemp_df, 'spread1')
        reg_contemp.compute
        puts "Contemporaneous regression (spread1 ~ spread2 + spread3):"
        puts reg_contemp.summary

        # --- Lagged Regression ---
        # Create lagged predictors: outcome at time t and predictors at time t-1.
        # To do that, drop the first row from outcome and last row from predictors.
        lagged_df = Daru::DataFrame.new({
            'spread1'     => Daru::Vector.new(spread1[1..-1]),
            'lag_spread2' => Daru::Vector.new(spread2[0...-1]),
            'lag_spread3' => Daru::Vector.new(spread3[0...-1])
        })
        reg_lag = Statsample::Regression::Multiple::Normal.new_from_dataset(lagged_df, 'spread1')
        reg_lag.compute
        puts "Lagged regression (spread1_t ~ spread2_(t-1) + spread3_(t-1)):"
        puts reg_lag.summary
    end
```
